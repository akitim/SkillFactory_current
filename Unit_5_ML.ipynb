Управляющий взял случайную выборку из 15 покупателей и выяснил, что их средние расходы равны 2000 рублей, а выборочное стандартное отклонение равно 400 рублей.

Управляющий хочет посчитать 95 %-ный доверительный интервал для истинного среднего значения расходов одного покупателя.

Какой из описанных выше функций следует воспользоваться управляющему для нахождения предела погрешности?


```python
from scipy.stats import t

def confidence_interval_t(alpha, s, n, mean):
    value = -t.ppf(alpha / 2, n - 1) * s / n**0.5
    return mean - value, mean + value

CL_ls = [90, 95, 99]
mu = 2000
s = 400
n = 15
for CL in CL_ls:
    alpha = 1 - CL / 100
    print(f'Доверительный интервал на {CL}% уровне доверия:',
          tuple([round(bound) for bound in confidence_interval_t(alpha, s, n, mu)])
    )
```

Обучите линейную регрессию из Задачи 3.7.3 при помощи градиентного спуска.
Чему равно оптимальное значение $\ L^2-loss$ ?

В модели линейной регрессии три наблюдения:
$$\ x_1 = 1, x_2 = 3, x_3 = 5$$
$$\ y_1 = 2.1, y_2 = 2.9, y_3 = 4.1 $$
$$\ y = w_0 + w_1 x $$

$$\ L = (2.1 − w_0 − w_1)^2 + (2.9 − w_0 − 3w_1)^2 + (4.1 − w_0 − 5w_1)^2 $$

```python
from scipy import optimize

def Loss(w):
    return (2.1 - w[0] - w[1]) ** 2 + (2.9 - w[0] - 3 * w[1]) ** 2 + (4.1 - w[0] - 5 * w[1]) ** 2

print ('Оптимальное значение L^2 - loss функции:', round(optimize.minimize(Loss, x0=(2, 1)).fun))
```

```python
import numpy as np
import pandas as pd
import numpy.linalg as npla
from sklearn.datasets import load_boston
from sklearn.metrics import mean_squared_error, f1_score, accuracy_score, roc_curve, roc_auc_score
from sklearn.model_selection import train_test_split

import warnings
warnings.filterwarnings('ignore')

def print_regression_metrics(y_true, y_pred):
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    print(f'MSE = {mse:.2f}, RMSE = {rmse:.2f}')
    
def prepare_boston_data():
    data = load_boston()
    X, y = data['data'], data['target']
    # Нормализовать даннные с помощью стандартной нормализации
    X = (X - X.mean(axis=0)) / X.std(axis=0)
    # Добавить фиктивный столбец единиц (bias линейной модели)
    X = np.hstack([np.ones(X.shape[0])[:, np.newaxis], X])
    
    return X, y

class RegOptimizer():
    def __init__(self, alpha, n_iters):
        self.theta = None
        self._alpha = alpha
        self._n_iters = n_iters
    
    def gradient_step(self, theta, theta_grad):
        return theta - self._alpha * theta_grad
    
    def grad_func(self, X, y, theta):
        raise NotImplementedError()

    def optimize(self, X, y, start_theta, n_iters):
        theta = start_theta.copy()

        # for i in range(n_iters):
        #     theta_grad = self.grad_func(X, y, theta)
        #     theta = self.gradient_step(theta, theta_grad)
        # tg_rec = []
        # theta_grad = self.grad_func(X, y, theta)
        # tg_rec += [theta_grad]

        threshold = 0.01
        i = 0

        while True:
            theta_grad = self.grad_func(X, y, theta)
            theta = self.gradient_step(theta, theta_grad)
            stop_cond = np.max(np.abs(theta_grad)) < threshold and i < n_iters
            if stop_cond: break
            i += 1

        # if i < n_iters:
        #     print(f'Остановлено по градиенту, {np.max(np.abs(theta_grad))}')
        # else:
        #     print(f'Остановлено по итерациям', {i})
        
        print(f'Вычисление остановилось на {i}-й итерации\nЗначение градиента theta:\n {theta_grad}' )
        return theta
    
    def fit(self, X, y):
        m = X.shape[1]
        start_theta = np.ones(m)
        self.theta = self.optimize(X, y, start_theta, self._n_iters)
        
    def predict(self, X):
        raise NotImplementedError()

class LinReg(RegOptimizer):
    def grad_func(self, X, y, theta):
        n = X.shape[0]
        grad = 1. / n * X.T @ (X @ theta - y)

        return grad
    
    def predict(self, X):
        if self.theta is None:
            raise Exception('You should train the model first')
        
        y_pred = X.dot(self.theta)
        
        return y_pred
```

```python
class LinRegAlgebra():
    def __init__(self):
        self.theta = None
    
    def fit(self, X, y):
        self.theta = npla.inv(X.T @ X) @ X.T @ y
    
    def predict(self, X):
        return X @ self.theta
```

$\ N$| 1 | 2 | 3 |
---:|:---:| --- | |
**Предсказанное значение** | 0.2 | 0 | 0.1
| 0.3 | 0 | 0   |
| 0.5 | 1 | 0.9 |
**Истинное значение** | 0 | 0 | 1
| 0 | 0 | 0 |
| 1 | 1 | 0 |



```python
import numpy as np

def logloss(y_t, y_p):
    def mgc_ln(x):
        return np.log(x) if x else x
    for i in range(y_p.shape[0]):
        for j in range(y_p.shape[1]):
            y_p[i, j] = mgc_ln(y_p[i, j])
    n = y_t.shape[1]
    return -1 / n *np.trace( y_t.T @ y_p)

y_p = np.array([[0.2, 0, 0.1], [0.3, 0, 0], [0.5, 1, 0.9]])
y_t = np.array([[0, 0, 1], [0, 0, 0], [1, 1, 0]])

print('Cредняя ошибка:', round(logloss(y_t, y_p), 4))
```

```python
adult['salary'] = (adult['salary'] != ' <=50K').astype('int32')
# Сделать one-hot encoding для некоторых признаков
adult = pd.get_dummies(adult, columns=['workclass', 'education', 'native_country',
                                       'marital-status', 'occupation', 'relationship',
                                       'race', 'sex'])
```

```python
# Нормализовать нуждающиеся в этом признаки
a_features = adult[['age', 'education-num', 'hours-per-week', 'fnlwgt', 'capital-gain', 'capital-loss']].values
norm_features = (a_features - a_features.mean(axis=0)) / a_features.std(axis=0)
adult.loc[:, ['age', 'education-num', 'hours-per-week', 'fnlwgt', 'capital-gain', 'capital-loss']] = norm_features
```

```python
# Разбить таблицу данных на матрицы X и y
X = adult[list(set(adult.columns) - set(['salary']))].values
y = adult['salary'].values

# Добавить фиктивный столбец единиц (bias линейной модели)
X = np.hstack([np.ones(X.shape[0]).reshape(-1, 1), X])
m = X.shape[1]
```

```python
lgrg4 = LogisticRegression(max_iter=1000)
lgrg4.fit(X, y)
y_p = lgrg4.predict(X)

print_logisitc_metrics(y, y_p)
```

Допустим, у вас есть выборки `y_true = [1.23, 2.35, 2.75]` и `y_pred = [1.01, 12.3, 2.74]`. Посчитайте метрику RMSE, округлите до сотых и запишите результат.

Загрузите данные `train.csv`, оставьте в данных только признаки `'fine_amount', 'state_fee', 'late_fee', 'discount_amount', 'balance_due'`, затем избавьтесь от пропусков. Целевой переменной будет `'balance_due'`, разделите данные на обучающую и тестовую выборки в соотношении 70% / 30% без перемешивания. Обучите линейную регрессию из `scikit-learn` и запишите в переменную `result` значение метрики RMSE на тестовой выборке. RMSE означает Rooted Mean Squared Error. Rooted означает, что из значения метрики был взят корень.

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

# path = '/content/gdrive/MyDrive/SkillFactory/train.csv'
# vis_data = pd.read_csv(path, encoding = 'ISO-8859-1', low_memory = False)

df = vis_data.loc[:, ['fine_amount', 'state_fee', 'late_fee', 'discount_amount', 'balance_due']]
print('Пропущенных значений в столбцах:\n', pd.isnull(df).sum())

df.dropna(axis=0, inplace=True)

X = df[['fine_amount', 'state_fee', 'late_fee', 'discount_amount']].values
y = df['balance_due'].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, shuffle=False)

lin_regr = LinearRegression()
lin_regr.fit(X_train, y_train)
y_pred = lin_regr.predict(X_test)
result = np.sqrt(mean_squared_error(y_test, y_pred))
print ('\nRMSE = ', result)
```

###Задание 4.6.6-4.6.10

Предположим, что у вас получилось создать модель-классификатор, которая не ошибается. Какое значение будет лежать вне главной диагонали в confusion matrix для предсказаний этой модели и истинных результатов?

```python
from sklearn.metrics import confusion_matrix
import numpy as np

y = np.random.randint(0, 2, size=5)
print('Матрица конфузии:\n', confusion_matrix(y, y))
```

На этапе вычисления метрик вы получили результаты` precision = 0.75` и `recall = 0.6`. Посчитайте значение метрики $\ f1$. Округлите до сотых и запишите результат.

```python
prec = 0.75
rec = 0.6
f1 = 2 * prec * rec / (prec + rec)
print ('F1-score:', round(f1, 2))
```

Допустим, у вас есть выборки `y_true = [0, 0, 1, 1, 1, 1, 0, 1]` и `y_pred = [0, 1, 0, 0, 1, 1, 0, 1]`. Посчитайте значение метрики $\ f1$. Округлите до сотых и запишите результат.

```python
y_true = np.array([0, 0, 1, 1, 1, 1, 0, 1])
y_pred = np.array([0, 1, 0, 0, 1, 1, 0, 1])

TP = np.count_nonzero(y_true * 2 + y_pred == 3)
FP = np.count_nonzero(y_true * 2 + y_pred == 1)
FN = np.count_nonzero(y_true * 2 + y_pred == 2)
precision = TP / (TP + FP)
recall = TP / (TP + FN)
f1 = 2 * precision * recall / (precision + recall)
print ('F1-score:', round(f1, 2))
```

Допустим, у вас есть выборки `y_true = [0, 0, 1, 0, 0, 1, 0]` и `y_pred = [1, 1, 1, 0, 1, 1, 0]`. Посчитайте значение метрики $\ precision $. Округлите до сотых и запишите результат.

```python
y_true = np.array([0, 0, 1, 0, 0, 1, 0])
y_pred = np.array([1, 1, 1, 0, 1, 1, 0])

TP = np.count_nonzero(y_true * 2 + y_pred == 3)
FP = np.count_nonzero(y_true * 2 + y_pred == 1)
precision = TP / (TP + FP)

print ('Precision:', round(precision, 2))
```

Допустим, у вас есть выборки `y_true = [0, 0, 1, 0, 0, 1, 0]` и `y_pred = [1, 1, 1, 0, 1, 1, 0]`. Посчитайте значение метрики  $\ recall $. Округлите до сотых и запишите результат.

```python
y_true = np.array([0, 0, 1, 0, 0, 1, 0])
y_pred = np.array([1, 1, 1, 0, 1, 1, 0])

TP = np.count_nonzero(y_true * 2 + y_pred == 3)
FN = np.count_nonzero(y_true * 2 + y_pred == 2)
recall = TP / (TP + FN)

print ('Recall:', round(recall, 2))
```
