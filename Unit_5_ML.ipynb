Управляющий супермаркета хочет оценить, сколько денег покупатели тратят за один поход в супермаркет в среднем. На основании ранее собранных данных известно, что расходы одного покупателя распределены приблизительно нормально.

Управляющий взял случайную выборку из 15 покупателей и выяснил, что их средние расходы равны 2000 рублей, а выборочное стандартное отклонение равно 400 рублей.

Управляющий хочет посчитать 95 %-ный доверительный интервал для истинного среднего значения расходов одного покупателя.

Какой из описанных выше функций следует воспользоваться управляющему для нахождения предела погрешности?


```python
from scipy.stats import t

def confidence_interval_t(alpha, s, n, mean):
    value = -t.ppf(alpha / 2, n - 1) * s / n**0.5
    return mean - value, mean + value

CL_ls = [90, 95, 99]
mu = 2000
s = 400
n = 15
for CL in CL_ls:
    alpha = 1 - CL / 100
    print(f'Доверительный интервал на {CL}% уровне доверия:',
          tuple([round(bound) for bound in confidence_interval_t(alpha, s, n, mu)])
    )
```

→ Необходимо найти 99 % доверительный интервал для истинного среднего значения налоговых выплат за год.

```python
def confidence_interval_norm(alpha, sigma, n, mean):
    value = -norm.ppf(alpha / 2) * sigma / n**0.5
    return mean - value, mean + value

mu = 3540
sigma = 1150
n = 250
CL = 99
alpha = 1 - CL / 100
print(f'Доверительный интервал на {CL}% уровне доверия:',\
      tuple([round(bound) for bound in confidence_interval_norm(alpha, sigma, n, mu)])
      )
```

#Задание 7.2

Обучите линейную регрессию из Задачи 3.7.3 при помощи градиентного спуска.
Чему равно оптимальное значение $\ L^2-loss$ ?

В модели линейной регрессии три наблюдения:
$$\ x_1 = 1, x_2 = 3, x_3 = 5$$
$$\ y_1 = 2.1, y_2 = 2.9, y_3 = 4.1 $$
$$\ y = w_0 + w_1 x $$

$$\ L = (2.1 − w_0 − w_1)^2 + (2.9 − w_0 − 3w_1)^2 + (4.1 − w_0 − 5w_1)^2 $$

```python
from scipy import optimize

def Loss(w):
    return (2.1 - w[0] - w[1]) ** 2 + (2.9 - w[0] - 3 * w[1]) ** 2 + (4.1 - w[0] - 5 * w[1]) ** 2

print ('Оптимальное значение L^2 - loss функции:', round(optimize.minimize(Loss, x0=(2, 1)).fun))
```
